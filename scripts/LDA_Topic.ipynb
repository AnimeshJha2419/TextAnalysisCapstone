```python
import nltk
from nltk.corpus import stopwords

# Define your custom stopwords
custom_stopwords = ['use','novel','br','human','science','work', 'also', 'past','witnessed','award','nsfs','merit','deemed','mission','worthy', 'decades','much','shown', 'non', 'project', 'research', 'systems', 'using', 'students', 'new', 'support','nsf','reflects','learning','based','high']

# Assuming 'df' is your DataFrame and 'Cleaned_Abstract' is the column containing text data
column_name = 'Abstract'  # Replace 'Cleaned_Abstract' with the actual column name

# Function to remove non-ASCII characters from a string
def _removeNonAscii(s):
    return "".join(i for i in s if ord(i) < 128)

# Apply the removal of non-ASCII characters and custom stopwords to each row in the specified column
for index, row in data_copy.iterrows():
    # Remove non-ASCII characters
    cleaned_text = _removeNonAscii(row[column_name])
    
    # Tokenize the text and remove stopwords (including custom stopwords)
    words = nltk.word_tokenize(cleaned_text)
    stopWords = set(stopwords.words('english') + custom_stopwords)  # Combine NLTK stopwords with custom stopwords
    filtered_words = [word for word in words if word.lower() not in stopWords]
    
    # Join the words back into a string and update the DataFrame
    cleaned_text = ' '.join(filtered_words)
    data_copy.at[index, column_name] = cleaned_text

# Display the updated DataFrame with non-ASCII characters removed and stopwords removed
print(data_copy)
```

```python
def get_keys(topic_matrix):
    '''
    returns an integer list of predicted topic 
    categories for a given topic matrix
    '''
    keys = topic_matrix.argmax(axis=1).tolist()
    return keys

def keys_to_counts(keys):
    '''
    returns a tuple of topic categories and their 
    accompanying magnitudes for a given list of keys
    '''
    count_pairs = Counter(keys).items()
    categories = [pair[0] for pair in count_pairs]
    counts = [pair[1] for pair in count_pairs]
    return (categories, counts)
```

```python
# Define helper functions
def get_top_n_words(n, keys, document_term_matrix, count_vectorizer):
    '''
    returns a list of n_topic strings, where each string contains the n most common 
    words in a predicted category, in order
    '''
    top_word_indices = []
    for topic in range(n_topics):
        temp_vector_sum = 0
        for i in range(len(keys)):
            if keys[i] == topic:
                temp_vector_sum += document_term_matrix[i]
        temp_vector_sum = temp_vector_sum.toarray()
        top_n_word_indices = np.flip(np.argsort(temp_vector_sum)[0][-n:],0)
        top_word_indices.append(top_n_word_indices)   
    top_words = []
    for topic in top_word_indices:
        topic_words = []
        for index in topic:
            temp_word_vector = np.zeros((1,document_term_matrix.shape[1]))
            temp_word_vector[:,index] = 1
            the_word = count_vectorizer.inverse_transform(temp_word_vector)[0][0]
            topic_words.append(the_word.encode('ascii').decode('utf-8'))
        top_words.append(" ".join(topic_words))         
    return top_words
```

```python
def get_mean_topic_vectors(keys, two_dim_vectors):
    '''
    returns a list of centroid vectors from each predicted topic category
    '''
    mean_topic_vectors = []
    for t in range(n_topics):
        articles_in_that_topic = []
        for i in range(len(keys)):
            if keys[i] == t:
                articles_in_that_topic.append(two_dim_vectors[i])    
        
        articles_in_that_topic = np.vstack(articles_in_that_topic)
        mean_article_in_that_topic = np.mean(articles_in_that_topic, axis=0)
        mean_topic_vectors.append(mean_article_in_that_topic)
    return mean_topic_vectors
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Assuming you have your LDA model, corpus, and id2word defined
# lda_model, corpus, id2word = ...

# Get topic-word distributions
topics = lda_model.show_topics(formatted=False)

# Create a graph
G = nx.Graph()

# Add edges between words within each topic based on their weights
for topic_id, words in topics:
    for word, weight in words:
        G.add_edge(f'Topic {topic_id}', word, weight=weight)

# Define node colors based on types (topic or word)
node_colors = [0 if 'Topic' in node else 1 for node in G.nodes]

# Draw the graph with improved layout
pos = nx.spring_layout(G, seed=42, iterations=100, k=0.2)
nx.draw_networkx_nodes(G, pos, nodelist=[node for node in G.nodes if 'Topic' in node], node_size=1000, node_color='lightblue', alpha=0.8)
nx.draw_networkx_nodes(G, pos, nodelist=[node for node in G.nodes if 'Topic' not in node], node_size=500, node_color='lightcoral', alpha=0.8)
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)

# Add labels for topics and words separately
topic_labels = {node: node.split()[1] for node in G.nodes if 'Topic' in node}
word_labels = {node: node for node in G.nodes if 'Topic' not in node}
nx.draw_networkx_labels(G, pos, labels=topic_labels, font_size=8, font_color='black', font_family='sans-serif', font_weight='bold')
nx.draw_networkx_labels(G, pos, labels=word_labels, font_size=6, font_color='black', font_family='sans-serif')

plt.title('Topic-Word Network Graph')
plt.show()
```
